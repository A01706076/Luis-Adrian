```python
df=pd.read_excel('datasets/Canada.xlsx',
                 engine='openpyxl',
                 sheet_name='Canada by Citizenship',
                 skiprows=range(20),
                 skipfooter=2
                )
df.head()
```

```python
df.drop(['Type','Coverage','AREA','REG','DEV',],
        axis=1,
        inplace=True
       )
#df.columns
```

```python
df.columns=list(map(str,df.columns))
df.columns
```

```python
df.set_index('Country',inplace=True)
df.head()
```

```python
df['Total']=df.iloc[:,
                    2:].sum(axis=1)
df.head(2)
```

```python
df.iloc[:,3]
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.style.use('ggplot')
```

```python
print(plt.style.available)
```

```python
df.loc['Japan','1980':'2013'].plot()
```

```python
ic_data=df.loc[['India','China'],'1980':'2013']
nic=ic_data.transpose()
nic.plot(figsize=(20,10))
plt.xlabel('Years',fontsize=17)
plt.xticks(fontsize=17, rotation=45)
plt.ylabel('Number of imigrants',fontsize=17)
plt.yticks(fontsize=17)
plt.title('Imigration to Canada from India and China during 1980 to 2013',fontsize=17)
```
