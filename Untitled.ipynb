```python
df=pd.read_excel('mic2.xlsx',
                 engine='openpyxl')
#df.head()
```

```python
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
mpl.style.use('ggplot')
```

```python
counts=df['Age'].value_counts()
counts
```

```python
counts=counts.reindex(['<=20','21 - 30','31 - 40','41 - 50'])
counts
```

```python
type(counts)
```

```python
counts=counts.to_frame()
```

```python
x=counts.loc[:,'Age'].sum(axis=0)
counts['%']=(counts.iloc[:,:]/x)*100
```

```python
counts
```

```python
counts.columns
```

```python
counts=counts.reset_index()
counts
```

```python
counts=counts.rename({'index':'Age_group'},
                     axis=1)
counts
```

```python
explode_list=[0.02,0.05,0.07,0.03]
color_list=['gold','yellowgreen','coral','skyblue','lightgreen','pink']
counts['Age'].plot(kind='pie',
                  figsize=(15,6),
                  autopct='%1.1f%%',
                  startangle=90,
                  shadow=True,
                   pctdistance=1.1,
                   labels=None,
                   explode=explode_list,
                   colors=color_list
                  )
```
